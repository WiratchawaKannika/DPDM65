```python
import pandas as pd
import numpy as np
```

# **Import Data**

```python
test_df = pd.read_csv('/content/drive/MyDrive/DPDM65/Final/Myfinaltest_X.csv')
print(test_df.shape)  ##บอกขนาดของ DataFrame
test_df.head()
```

### Dummy variable

```python
colName = ['property_type',	'furniture', 'number_of_story',	'number_of_bedroom',	'number_of_bathroom',
              	'number_of_parking',	'central_cost',	'parking_percentage',	'area',	'usable_area']
testset = test_df[colName]
testset.head()

# Dummy variable
categorical = ['property_type', 'furniture']
df_encode = pd.get_dummies(data = testset,  columns = categorical, drop_first =True)
print(df_encode.shape)
df_encode.head()
```

## Prediction 

#### Loading the scikit learn models with Pickle

```python
# Loading the saved decision tree model pickle
import pickle

model_filename = '/content/drive/MyDrive/DPDM65/Final/DecisionTreeRegress_final.pkl'
dec_tree_regre_pkl = open(model_filename, 'rb')
dec_tree_regre_model = pickle.load(dec_tree_regre_pkl)
print("Loaded Decision tree model :: ", dec_tree_regre_model)
```

```python
# prediction from Load model
pred = dec_tree_regre_model.predict(df_encode)
print(len(pred))
print(pred)
```

## Save To CSV.

```python
list_pred = pred.tolist()
print(len(list_pred))
```

```python
df_pred = test_df[['INDEX']]
df_pred['sell_price'] = list_pred 
df_pred = df_pred.sort_values(by='INDEX', ascending=True)
print(df_pred.shape)
df_pred.head()
```
