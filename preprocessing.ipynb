### pandas.DataFrame.dropna

`DataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)`

- axis (default = 0) ต้องการสนใจในแนว rows หรือ columns
- thresh แถวนั้นจะถูก drop ถ้ามีจำนวน Na น้อยกว่าจำนวนที่เรา set ไว้ 
- subset คือ ระบุ columns ที่ต้องการ dropna() เป็น list ของ Colums เช่น   ['colums1', 'columns2'] 
- inplace คือ function ที่ keep ผลของการ dropna() ไว้ในตารางนั้นเลย โดยไม่ต้องสร้างตัวแปรขึ้นมารับค่าไว้ใหม่ คือ การ update ตัวแปรเดิม หรือ ใช้เมื่อจะไม่นำตารางเดิมไปใช้ต่อ

````python
x1_, x2_, x3_, x4_, x5_, fac_, squre_, div_ = [],[],[],[],[],[],[],[]
#ourNum = range(1,6)
ourRange = range(1,16)
#for n in ourNum:
for x in ourRange:
    x2 = 2 * x
    x3 = 3 * x
    x4 = 4 * x
    x5 = 5 * x
    fac = np.math.factorial(x)
    squre = x ** 2
    if squre % 3 == 0:
        div_.append('yes')
    else:
        div_.append('no')
    #print(n," * ",x," = ",result)
    x1_.append(x)
    x2_.append(x2)
    x3_.append(x3)
    x4_.append(x4)
    x5_.append(x5)
    fac_.append(fac)
    squre_.append(squre)

Table_xN = {'x1':x1_,'x2':x2_,'x3':x3_,'x4':x4_,'x5':x5_, 'xN': xN_, 'Factorial':fac_, 'Squre': squre_,'Divisible': div_}
Table_xN_ = pd.DataFrame(Table_xN, columns=['x1','x2','x3','x4','x5', 'Factorial', 'Squre','Divisible'])
print(Table_xN_.shape)
Table_xN_
````
